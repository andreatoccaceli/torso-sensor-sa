### 0D lumped-parameter model plots

Idea: $Q_{\text{AV}} - Q^{\text{SYS}}_{\text{AR}}$ is a proxy of the PPG signal measured. Thus we could directly link an observable to a state of our dynamical system. If we could find an anchor value (i.e. the value of a state variable in a time istant) then the system could be calibrated matching the observable with the state variable

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot ionic solver variables
DIR_PATH = "/home/andrea/Scrivania/phd/dev/torso-sensor-sa/0D-volumes/"
DB_PATH = f"{DIR_PATH}/circulation_regurgitant_reduced.csv"
DB_2_PATH = f"{DIR_PATH}/circulation_reduced.csv"

db = pd.read_csv(DB_PATH)
db_2 = pd.read_csv(DB_2_PATH)

db.plot(x="time", y="pVENSYS", xlim = (78, 80), grid=True)#xlim=(98.4, 100.0), grid=True)
db_2.plot(x="time", y="pVENSYS", xlim = (78, 80), grid=True)#xlim=(98.4, 100.0), grid=True)

plt.show()
```

```python
raw_times = db['time']
len(raw_times)
```

```python
ig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: pLV - pARSYS
y_reg = db["pLV"] - db["pARSYS"]
y_base = db_2["pLV"] - db_2["pARSYS"]

ax1.plot(db_2["time"], y_base, color='tab:blue', linewidth=2.5, label='pLV - pARSYS (Base)')
ax1.plot(db["time"], y_reg, color='darkred', linewidth=2.5, linestyle='--', label='pLV - pARSYS (Regurgitant)')
ax1.grid(True, linestyle=':', alpha=0.7)
ax1.set_xlim(78, 80)
ax1.legend(loc='upper right')

# Plot 2: pLV
ax2.plot(db_2["time"], db_2["pLV"], color='tab:blue', linewidth=2.5, label='pLV (Base)')
ax2.plot(db["time"], db["pLV"], color='darkred', linewidth=2.5, linestyle='--', label='pLV (Regurgitant)')
ax2.grid(True, linestyle=':', alpha=0.7)
ax2.set_xlim(78, 80)
ax2.legend(loc='upper right')

# Plot 3: QAV
ax3.plot(db_2["time"], db_2["QAV"], color='tab:blue', linewidth=2.5, label='QAV (Base)')
ax3.plot(db["time"], db["QAV"], color='darkred', linewidth=2.5, linestyle='--', label='QAV (Regurgitant)')
ax3.grid(True, linestyle=':', alpha=0.7)
ax3.set_xlim(78, 80)
ax3.legend(loc='upper right')
```

```python
fig2, ax4 = plt.subplots(1, 1, figsize=(18, 5))

ax4.plot(db_2["time"], db_2["VLV"], color='tab:blue', linewidth=2.5, label='VLV (Base)')
ax4.plot(db["time"], db["VLV"], color='darkred', linewidth=2.5, linestyle='--', label='VLV (Regurgitant)')
ax4.grid(True, linestyle=':', alpha=0.7)
ax4.set_xlim(78, 80)
ax4.legend(loc='upper right')

plt.show()
```
